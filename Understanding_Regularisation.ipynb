```
import pandas as pd
import numpy as np 
import seaborn as sns
import matplotlib.pyplot as plt
import random
import math
```

```
T = 50
```

```
# coeffs = np.random.uniform(-0.5,0.5,4)
coeffs = np.array([1.2341, 0.29932241, -0.11951394,  0.31870231, -0.05287935])
ts = np.linspace(0,5,T)
t_mat = np.stack([np.ones_like(ts),ts,ts**2,ts**3,ts**4]).T
fn = (coeffs*t_mat).sum(axis=1)
```

```
plt.plot(fn)
```

```
defects = np.random.poisson(fn)
```

```
plt.plot(ts,fn)
plt.scatter(ts,defects,s=3,c="tab:orange")
```

```
plt.plot(ts,fn)
plt.scatter(ts,defects,s=3,c="tab:orange")
plt.plot(ts,defects)
```

```
fact = np.vectorize(math.factorial)
# Optimization Function -
def loss_fn(lam, x):
    # Negative Likelihood Loss function + Regularisation
    terms = -(lam + x*np.log(lam) + np.log(fact(x)))
    return np.sum(terms)

def loss_grad(lam, x,alpha=0.1):
    # NLL Gradient 
    reg_loss = np.zeros_like(lam).astype(np.float32)
    for i in range(1,len(lam)):
        reg_loss[i]= alpha*(lam[i] - lam[i-1])

    #print(reg_loss)

    return 1- x/(lam) +reg_loss

```

```
lambda_i = np.random.uniform(0.5,1.5,size=(len(defects)))
ALPHA = 1
EPOCHS = 100
lr = 0.05
losses = []
# Gradient Descent
for i in range(EPOCHS):
    losses.append(1)
    lambda_i -= lr*loss_grad(lambda_i,defects,ALPHA)

print(lambda_i)
plt.plot(ts,fn)
plt.scatter(ts,defects,s=3,c="tab:orange")
plt.plot(ts,lambda_i)
```
